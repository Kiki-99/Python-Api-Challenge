```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import weather_api_key
```

```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
```

```python
#Installing Citipy Module
from citipy import citipy

#Extracting response for 500 cities
cities = []
city_weather_data =[]
lats = np.random.randint(low=-90, high=90, size=1000)
lons = np.random.randint(low=-180, high=180, size=1000)
for lat, lon in zip(lats, lons):
    city = citipy.nearest_city(lat, lon)
    #print(city.city_name)
    
    city_weather_request_url = url + "appid=" + weather_api_key + "&q=" + city.city_name
    print(city_weather_request_url)
    
    response = requests.get(city_weather_request_url)
    if response.status_code !=404:
        city_weather_data.append(response.json())
        
    elif response.status_code ==500:
        print(city_weather_request_url)
```

```python
city_weather_data[0]
```

```python
# List for holding temperature, humidity, cloudiness and windspeed
# Extract interesting data from responses
    
extracted_data = []
lat_data = 0
temp_data =0
for city_weather in city_weather_data:
    lat_data = city_weather.get("coord").get("lat")
    lon_data = city_weather.get("coord").get("lon")
    temp_data = city_weather.get("main").get("temp_max")
    humidity_data = city_weather.get("main").get('humidity')
    cloud_data = city_weather.get("clouds").get("all")
    wind_speed_data = city_weather.get("wind").get("speed")
    country_data = city_weather.get("sys").get("country")
    date_data = city_weather.get("dt")
    
    extract = {'City': city_weather.get('name'),"Max Temp": temp_data, "Lat": lat_data, "Lng": lon_data, "Humidity": humidity_data, "Cloudiness": cloud_data, "Wind Speed": wind_speed_data, "Country": country_data, "Date": date_data}
    extracted_data.append(extract)

city_weather_data_df = pd.DataFrame(extracted_data)
city_weather_data_df
```

```python
#Saved data to CSV

city_weather_data_df.to_csv("Weather Data")
```

```python
from scipy.stats import linregress

#Calculate citites in the Northen Hemisphere
Northern_Hem_df = city_weather_data_df.loc[city_weather_data_df["Lat"] >=0]
Northern_Hem_df

#Create Linear Regression for Northern Hemisphere Temp vs. Latitude
x_values = Northern_Hem_df["Lat"]
y_values = Northern_Hem_df["Max Temp"]
slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.scatter(Northern_Hem_df["Lat"], Northern_Hem_df["Max Temp"])
plt.show()
```
